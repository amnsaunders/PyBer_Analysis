```python
# Import Numpy for calculations and matplotlib for charting
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a numpy array from 0 to 5 with each step being 0.1 higher than the last
x = np.arange(0, 5, 0.1)
x
```

```python
# creates an exponential series of values which we can then chart
e_x = [np.exp(x) for x in x]
e_x
```

```python
#create a graph based upon the list and array we have created
plt.plot(x,e_x)
# Show the graph
plt.show()
```

```python
# give graph axis labels
plt.xlabel("Time")
plt.ylabel("Increments")
plt.plot(x,e_x)
plt.show()
```
